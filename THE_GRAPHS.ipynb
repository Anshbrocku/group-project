```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('Data/credit3.csv')

data = data.loc[:,['fraud','used_pin_number',"used_chip","online_order"]]

noPin = 0
pin = 0

index = data['fraud'] == 1
FraudPins = data.loc[index, 'used_pin_number']

for i in FraudPins:
    if i == 0:
        noPin += 1
    else:
        pin += 1

x = ["fraud that used pins", "fraud that didnt use pins"]
y = [pin,noPin]

#print(noPin, "  ",pin)

percent = pin/len(FraudPins)*100

#print("\n\nthe percent of fraud cases that use pin are: ", str(percent) + "%")


plt.bar(x,y)
plt.show()
```
